# **Importação das Bibliotecas**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# **Importação da Base de Dado**

```python
videos_df = pd.read_excel('/content/drive/MyDrive/Data python 2413/05- Exercícios de Seaborn/videosYT.xlsx', 'videos')#Importa a parte dos videos se tiver mais uma planilha
display(videos_df)
```

# **Criação de Gráfico de Dispersão**

```python
fig = sns.scatterplot(data = videos_df, x = 'Nº de Views', y = 'Nº de Likes',hue = 'Responsável', style = 'Responsável', palette = ['red','black'])
plt.show(fig)
```
