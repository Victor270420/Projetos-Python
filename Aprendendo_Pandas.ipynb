# **Passo_9.: Excluir Colunas e Linhas.**

```python
vendas_df = vendas_df.drop('Imposto',axis=1)#axis=1 é para apagar a coluna e axis=0 exclui linha
display(vendas_df)
```

```python
vendas_df = vendas_df.drop(0,axis = 0)#axis=1 é para apagar a coluna e axis=0 exclui linha
display(vendas_df)
```

# **Passo_10.: Tratamento e analise de Dados.**

# **Passo_11.: Calcular indicadores.**

```python
transacoes_loja = vendas_df['ID Loja'].value_counts()
display(transacoes_loja)
```

```python
faturamento_produto = vendas_df[['Produto','Valor Final','Quantidade']].groupby('Produto').sum()#Soma os Produtos so os valores finais e quantidades dos produtos
display(faturamento_produto)
```

```python
faturamento_produto = vendas_df[['Produto','Valor Final','Quantidade']].groupby('Produto').mean()#Média dos produtos cada vez que passa em um produto soma e no final calcula a média
display(faturamento_produto)
```
