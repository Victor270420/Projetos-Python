```python
import plotly.express as px
from google.colab import output
lista_y = []
lista_x = []

quantidade =  int(input('Digite a quantidades de anos '))

#Cloleta dos dados do grafico

for quant in range(1,quantidade+1):
  x = str(input("Digite o ano da venda: "))
  lista_x.append(x)

  vendas_y = int(input(f"Digite o número de vendas em {x}: "))
  lista_y.append(vendas_y)

#Criando as variaves para a edição para frente

nome_titulo = input('Digite o nome do titulo')
altura = 500
largura = 1200
tabela_estilo = px.line
linha_estilo = 'solid'
titulo_eixo_Y = input('Digite o titulo do eixo Y')
titulo_eixo_X = input('Digite o titulo do eixo X')
tick_estilo = 'inside'
cores_geral=['red','blue','green','yellow','orange','purple','pink','brown','black','gray']
cor_x_titulo_principal = 'black'
cor_y_titulo_principal = 'black'
cor_y_titulo_secundario = cor_y_titulo_principal
cor_x_titulo_secundario = cor_x_titulo_principal
cor_tiulo = 'red'
cor_linha = 'red'
output.clear()
#Criando graficos e tabelas
fig = tabela_estilo( x = lista_x , y = lista_y, title= nome_titulo, height = altura, width = largura , line_shape ='spline')
fig.update_layout(title_font_color = cor_tiulo)#titulo
fig.update_traces(line = dict(color = cor_linha, dash=linha_estilo))#Traço da tabela dash tipo da linha
fig.update_xaxes(title = titulo_eixo_X , title_font_color = cor_x_titulo_principal , ticks = tick_estilo , tickfont_color = cor_x_titulo_secundario)#eixo X
fig.update_yaxes(title = titulo_eixo_Y , title_font_color = cor_y_titulo_principal , ticks = tick_estilo , tickfont_color = cor_y_titulo_secundario)#eixo y
fig.show()

while True:
    escolha = int(input('''Digite quais aletraçoes deseja:
    1 para alterar o tamanho da tabela,
    2 para alterar o eixo Y ou eixo X,
    3 para alterar o titulo,
    4 para alterar a linha,
    5 para mudar os dados da tabela,
    6 para ver a tabela,
    7 para saír'''))
    output.clear()
    while escolha == 1:#Funcionando
      tamanho = input('''Digite o tamnho que deseja:
      Pequeno (P)
      Médio (M)
      Grande (G)
      para saír digite (S)
      ''').title()
      output.clear()
      if tamanho == 'P':
        altura = 300
        largura = 600
        print('Alteração feita')
      elif tamanho == 'M':
        altura = 400
        largura = 1200
        print('Alteração feita')
      elif tamanho == 'G':
        altura = 600
        largura = 1800
        print('Alteração feita')
      elif tamanho == 'S':
        break
      else:
        print('Escolha errada tente novamente:')
    if escolha == 2:#Fazer o eixo X e Y #Funcionando
      eixo = int(input('''Digite o eixo que deseja altera:
      1 para alterar o X
      2 para alterar o Y
      '''))
      output.clear()
      while eixo == 1:
        alteracao_eixo_X = input('''Digite a alteração que deseja fazer:
        Para colocar um titulo (T)
        Cor dos textos e titulos (C)
        Colocar tiks (CT)
        para saír digite (S)
        ''').upper()
        output.clear()
        if alteracao_eixo_X == 'T':
          titulo_eixo_X = input('Digite o titulo do eixo X')
        elif alteracao_eixo_X == 'C':
          escolha_titulo_cor_x = int(input('''Qual alteração desaeja fazer:
          1 para cor dos textos
          2 para cor do titulo principal do eixo
          '''))
          if escolha_titulo_cor_x == 1:
            cor_texto_eixo_X = int(input('''Catalogo de cores:
                    1: red
                    2: blue
                    3: green
                    4: yellow
                    5: orange
                    6: purple
                    7: pink
                    8: brown
                    9: black
                    10: gray'''))
            for indice,cor in enumerate((cores_geral)):
              if cor_texto_eixo_X == indice+1:
                cor_x_titulo_secundario = str(cor)
              else:
                continue

          elif escolha_titulo_cor_x == 2:
            cor_texto_eixo_X = int(input('''Catalogo de cores:
                    1: red
                    2: blue
                    3: green
                    4: yellow
                    5: orange
                    6: purple
                    7: pink
                    8: brown
                    9: black
                    10: gray'''))
            for indice,cor in enumerate((cores_geral)):
              if cor_texto_eixo_X == indice+1:
                cor_x_titulo_principal = str(cor)
              else:
                continue
        elif alteracao_eixo_X == 'CT':
          tick_escolha = int(input('''Catogo de tiks:
          1 para coloca os ticks do eixo para dentro do gráfico.
          2 para coloca os ticks do eixo para fora do gráfico.
          3 para defalte'''))
          output.clear()
          if tick_escolha == 1:
            tick_estilo = 'inside'
            print('Alteração feita')
          elif tick_escolha == 2:
            tick_estilo = 'outside'
            print('Alteração feita')
          elif tick_escolha == 3:
            tick_escolha = 'default'
        elif alteracao_eixo_X == 'S':
          break
      while eixo == 2:
        alteracao_eixo_Y = input('''Digite a alteração que deseja fazer:
        Para colocar um titulo (T)
        Cor dos textos e titulos (C)
        Colocar tiks (CT)
        para saír digite (S)
        ''').upper()
        output.clear()
        if alteracao_eixo_Y == 'T':
          titulo_eixo_Y = input('Digite o titulo do eixo Y')
        elif alteracao_eixo_Y == 'C':
          escolha_titulo_cor_y = int(input('''Qual alteração desaeja fazer:
          1 para cor dos textos
          2 para cor do titulo principal do eixo '''))
          if escolha_titulo_cor_y == 1:
            cor_texto_eixo_X = int(input('''Catalogo de cores:
                    1: red
                    2: blue
                    3: green
                    4: yellow
                    5: orange
                    6: purple
                    7: pink
                    8: brown
                    9: black
                    10: gray'''))
            for indice,cor in enumerate((cores_geral)):
              if cor_texto_eixo_X == indice+1:
                cor_y_titulo_secundario = str(cor)
              else:
                continue

          elif escolha_titulo_cor_y == 2:
            cor_texto_eixo_X = int(input('''Catalogo de cores:
                    1: red
                    2: blue
                    3: green
                    4: yellow
                    5: orange
                    6: purple
                    7: pink
                    8: brown
                    9: black
                    10: gray'''))
            for indice,cor in enumerate((cores_geral)):
              if cor_texto_eixo_X == indice+1:
                cor_y_titulo_principal = str(cor)
              else:
                continue
        elif alteracao_eixo_Y == 'CT':
          tick_escolha = int(input('''Catogo de tiks:
          1 para coloca os ticks do eixo para dentro do gráfico.
          2 para coloca os ticks do eixo para fora do gráfico.
          3 para defalte'''))
          output.clear()
          if tick_escolha == 1:
            tick_estilo = 'inside'
            print('Alteração feita')
          elif tick_escolha == 2:
            tick_estilo = 'outside'
            print('Alteração feita')
          elif tick_escolha == 3:
            tick_escolha = 'default'
        elif alteracao_eixo_Y == 'S':
          break
    while escolha == 3:#Funcionando
      titulo_escolha = int(input('''Digite o eixo que deseja altera:
      1 alterar cor do titulo
      2 alterar titulo
      3 sair
      '''))
      output.clear()
      if titulo_escolha == 1:
        cor_texto_eixo_X = int(input('''Catalogo de cores:
                    1: red
                    2: blue
                    3: green
                    4: yellow
                    5: orange
                    6: purple
                    7: pink
                    8: brown
                    9: black
                    10: gray'''))
        output.clear()
        for indice,cor in enumerate((cores_geral)):
          if cor_texto_eixo_X == indice+1:
            cor_tiulo = str(cor)
          else:
            continue
      elif titulo_escolha == 2:
        nome_titulo = input('Digite o novo nomme do titulo: ')
      elif titulo_escolha == 3:
        break
      else:
        print('Erro tente novamente')
    while escolha == 4:#Funcionando
      escolha_linha =int(input('''Digite se quer modificar:
      1 cor da linha
      2 formato da linha
      3 para sair'''))
      if escolha_linha ==1:
        cor_texto_eixo_X = int(input('''Catalogo de cores:
        1: red
        2: blue
        3: green
        4: yellow
        5: orange
        6: purple
        7: pink
        8: brown
        9: black
        10: gray'''))
        output.clear()
        for indice,cor in enumerate((cores_geral)):
          if cor_texto_eixo_X == indice+1:
            cor_linha = str(cor)
          else:
            continue
      elif escolha_linha == 2:
        linha_escolha = int(input('''Digite o tipo de linha que deseja:
        1 Linha contínua.
        2 Linha tracejada (curtas).
        3 Linha composta por pontos.
        4 Linha com um padrão alternado de traços e pontos
        6 Para sair
        '''))
        if linha_escolha == 1:
          linha_estilo = 'solid'
          print('Alteração feita')
        elif linha_escolha == 2:
          linha_estilo = 'dash'
          print('Alteração feita')
        elif linha_escolha == 3:
          linha_estilo = 'dot'
          print('Alteração feita')
        elif linha_escolha == 4:
          linha_estilo = 'dashdot'
          print('Alteração feita')
        elif linha_escolha == 5:
          break
      elif escolha_linha == 3:
        break
    if escolha == 5:
      lista_y = []
      lista_x = []
      quantidade =  int(input('Digite a quantidades de anos '))
      for quante in range(1,quantidade+1):
         x = str(input("Digite o ano da venda: "))
         lista_x.append(x)

         vendas_y = int(input(f"Digite o número de vendas em {x}: "))
         lista_y.append(vendas_y)
    elif escolha == 6:#tabela
      fig = tabela_estilo(x = lista_x , y = lista_y , height = altura , width = largura , title= nome_titulo , line_shape ='spline')
      fig.update_layout(title_font_color = cor_tiulo)#titulo
      fig.update_traces(line = dict(color = cor_linha , dash = linha_estilo))#linha
      fig.update_xaxes(title = titulo_eixo_X , title_font_color = cor_x_titulo_principal , ticks = tick_estilo , tickfont_color = cor_x_titulo_secundario)#eixo X
      fig.update_yaxes(title = titulo_eixo_Y , title_font_color = cor_y_titulo_principal , ticks = tick_estilo , tickfont_color = cor_y_titulo_secundario)#eixo y
      fig.show()#tabela

    elif escolha == 7:#Fim do codigo
      print(40*'-')
      print('FIM')
      print(40*'-')
      break
```
