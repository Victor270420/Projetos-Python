# **Importação da Biblioteca**

```python
import matplotlib.pyplot as plt
```

# **Criação do grafico com Plot()**

```python
x = [1, 2, 3, 4]
y = [2,3,4,3]

plt.plot(x,y)
plt.show()
```

# **Alteração no Grafico**

```python
x = [1, 2, 3, 4]
y = [2,3,4,3]

plt.plot(x,y,label = 'dados')
plt.legend()#Coloca a cor da linha e o titulo no canto superior direito
plt.title('Grafico de Linha', color = 'red')#Coloca o titulo
plt.xlabel('Eixo X',color = 'red')#Titulo eixor X
plt.ylabel('Eixo Y',color = 'red')#Titulo eixo Y
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,3,4])
plt.show()#Mostra o grafico
```

# **Propriedades da Linha**

```python
x = [1, 2, 3, 4]
y = [2,3,4,3]


plt.plot(x,y, linestyle = '--',color = 'b', lw = 3.0, marker = 'o', label = 'Dados')
plt.xlabel('Dias', color = 'red')
plt.ylabel('Vendas', color = 'red')
plt.title('Grafico de Vendas', color = 'red')

#'upper right': canto superior direito
#'upper left': canto superior esquerdo
#'lower left': canto inferior esquerdo
#'lower right': canto inferior direito
#'right': canto direito
#'center left': centro esquerdo
#'center right': centro direito
#'lower center': centro inferior
#'upper center': centro superior
#'center': centro

#'b' Azul (Blue)
#'g' verde(Green)
#'r' Vermelho (Red)
#'c' Ciano (Cyan)
#'m' Magenta (Magenta)
#'y' Amarelo (Yellow)
#'k' Preto (Black)
#'w' Branco (White)

#Os marcadores podem ser: o,^,v,>,<,D,*,+,x,.,p e h

#plt.plot(x,y, linestyle = '-')#Deixa a linha solido
#plt.plot(x,y, linestyle = ':')#Deixa a linha pontilhado
#plt.plot(x,y, linestyle = '-.')#Deixa a Linha tracejada e pontilhada
plt.legend(loc = 'upper left')
plt.show()
```

# **Criação um Gráfico de Dispersão Scatter()**

```python
x = [1, 2, 3, 4]
y = [2,3,4,3]
plt.scatter(x,y, label = 'Dados')
plt.xlabel('Eixo X', color = 'red')
plt.ylabel('Eixo Y', color = 'red')
plt.title('Grafico de Dispersão', color = 'red')
plt.legend()
```

# **Criação de Gráfico de Barras**

```python
x = [1, 2, 3, 4]
y = [2,3,4,3]

plt.bar(x,y, label = 'Dados', color = 'red')
plt.xlabel('Eixo X', color = 'red')
plt.ylabel('Eixo Y', color = 'red')
plt.title('Grafico de Barras', color = 'red')
plt.legend()
plt.show()
```

# **Mudando os Tipos de Marcadores**

```python
x = [1, 2, 3, 4]
y = [2,3,4,3]

plt.scatter(x,y, color = 'k', marker = 's' )

#Os marcadores podem ser: o,^,v,>,<,D,*,+,x,.,p e h

plt.show()
```

# **Alterando os Limites dos Eixos**

```python
x = [1, 2, 3, 4]
y = [2,3,4,3]

plt.plot(x,y)
plt.axis(xmin = -1, xmax = 20, ymin = -5 , ymax = 15)#Alterando os limites dos eixos
plt.show()
```

```python
plt.plot(x,y)
plt.axis('auto')#Melhor visualização
plt.show()
```

```python
plt.plot(x,y)
plt.axis('square')#Deixa quadrado
plt.show()
```

# **Criação de Figure e Subplots()**

```python
valores_x = [1, 2, 3, 4]
valores_y = [1,4,2,3]

figura = plt.figure(figsize = (30,4))
figura.suptitle('Análise Geral')
facecolor = 'blue'

#Primiro Gráfico da Figura
figura.add_subplot(1,3,1)#1 primeia linha 3 quantos graficos vai ter 1 primeiro grafico
plt.plot(valores_x, valores_y,label = 'Dados')
plt.legend()
plt.title('Gráfico 1')

#Terceiro Gráfico da figura
figura.add_subplot(1,3,3)
plt.scatter(valores_x, valores_y, label = 'Dados')
plt.legend()
plt.title('Gáfico 3')

#Segundo Gráfico
figura.add_subplot(1,3,2)
plt.bar(valores_x, valores_y, label = 'Dados')
plt.legend()
plt.title('Gráfico 2')

plt.savefig('grafico.png', dpi = 100)
plt.show(figura)
```

# **Importção da Base de Dados - Bolsa de Valores**

```python
import pandas as pd
cotacao_df = pd.read_csv('/content/drive/MyDrive/Data python 2413/04- Exercício de Matplotlib/indexData.csv')
display(cotacao_df)
```

# **Criando um Novo DataFrame com 3 Colunas**

```python
cotacao_df1 = cotacao_df[['Index','Date','Close']]
display(cotacao_df1)
```

# **Reduzindo para Apenas NYA**

```python
dfremove = cotacao_df1.loc[(cotacao_df1['Index']!='NYA')]#Cria uma df diferente de NYA
cotacao_dffinal = cotacao_df1.drop(dfremove.index)#Apga tudo que for diferente de NYA
display(cotacao_dffinal)
```

```python
cotacao_dffinal = cotacao_dffinal[13900:]
display(cotacao_dffinal)
```

# **Criando um Gráfico de Linhas com a base de Nova York**

```python
plt.figure(figsize = (20,4))
plt.plot(cotacao_dffinal['Date'], cotacao_dffinal['Close'], label = 'NYA', color = 'r', ls = '--', lw='2')
plt.legend(loc = 2, fontsize = 18)
plt.axis(xmin = '2021-03-01', xmax = '2021-03-23')
plt.xlim(['2021-02-01', '2021-03-23'])
plt.xticks(rotation = 45, fontsize = 10)
plt.ylabel('Volume do Fechamento')
plt.xlabel('Data da Cotação')
plt.title('Cotação do Indice NYA')

plt.show()
```
